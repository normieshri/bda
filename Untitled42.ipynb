```python
# Install Java
!apt-get install openjdk-8-jdk-headless -qq > /dev/null

# Download and extract Hadoop
!wget -q https://archive.apache.org/dist/hadoop/common/hadoop-3.3.0/hadoop-3.3.0.tar.gz
!tar xf hadoop-3.3.0.tar.gz
!mv hadoop-3.3.0 /opt/hadoop

# Set up Hadoop environment variables
import os

os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"
os.environ["HADOOP_HOME"] = "/opt/hadoop"
os.environ["HADOOP_CONF_DIR"] = "/opt/hadoop/etc/hadoop"
os.environ["PATH"] = os.environ["PATH"] + ":/opt/hadoop/bin"

# Create necessary Hadoop directories
!mkdir -p /opt/hadoop_store/hdfs/namenode
!mkdir -p /opt/hadoop_store/hdfs/datanode

# Configure Hadoop
!echo "export JAVA_HOME=/usr/lib/jvm/java-8-openjdk-amd64" >> /opt/hadoop/etc/hadoop/hadoop-env.sh
!echo "export HDFS_NAMENODE_USER=root" >> /opt/hadoop/etc/hadoop/hadoop-env.sh
!echo "export HDFS_DATANODE_USER=root" >> /opt/hadoop/etc/hadoop/hadoop-env.sh
!echo "export HDFS_SECONDARYNAMENODE_USER=root" >> /opt/hadoop/etc/hadoop/hadoop-env.sh
!echo "export YARN_RESOURCEMANAGER_USER=root" >> /opt/hadoop/etc/hadoop/yarn-env.sh
!echo "export YARN_NODEMANAGER_USER=root" >> /opt/hadoop/etc/hadoop/yarn-env.sh

# Format HDFS
!hdfs namenode -format -force

# Start Hadoop services
!/opt/hadoop/sbin/start-dfs.sh
!/opt/hadoop/sbin/start-yarn.sh

# Check Hadoop status
!jps
```

```python
# Install Ant (if not already installed)
!apt-get install ant -qq > /dev/null

# Download Pig source code
!wget -q https://archive.apache.org/dist/pig/pig-0.17.0/pig-0.17.0-src.tar.gz
!tar xf pig-0.17.0-src.tar.gz

# Remove existing destination directory
!rm -r /opt/hadoop/pig/pig-0.17.0-src

# Move Pig source code to destination
!mv pig-0.17.0-src /opt/hadoop/pig

# Manually download Ivy jar and place it in the Pig directory
!wget -q -O /opt/hadoop/pig/ivy/ivy-2.2.0.jar http://archive.apache.org/dist/ant/ivy-2.2.0/ivy-2.2.0.jar

# Build Pig using Ant
!cd /opt/hadoop/pig && ant

# Move generated JAR files to Pig directory
!mv /opt/hadoop/pig/build/*.jar /opt/hadoop/pig/lib

# Set Pig environment variables
import os

os.environ["PIG_HOME"] = "/opt/hadoop/pig"
os.environ["PATH"] = os.environ["PATH"] + ":/opt/hadoop/pig/bin"
os.environ["PIG_CLASSPATH"] = os.environ["HADOOP_HOME"] + "/conf"

# Print Pig version
!pig -version

```
